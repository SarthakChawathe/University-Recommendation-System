```python
valid_majors = major_counts.index[:25]

# create a dictionary to map misspelled/abbreviated major names to their corrected versions
major_dict = {}
for major in top_accepted_df['Major'].unique():
    if major not in valid_majors:
        match = process.extractOne(major, valid_majors, scorer=fuzz.token_set_ratio)
        if match[1] >= 70:  # adjust the threshold as needed
            major_dict[major] = match[0]

# replace the misspelled/abbreviated major names with their corrected versions
top_accepted_df.loc[:, 'Major'] = top_accepted_df['Major'].replace(major_dict)

# keep only the rows with valid major names
top_accepted_df = top_accepted_df.loc[top_accepted_df['Major'].isin(valid_majors)]
```

```python
data = top_accepted_df.copy()
toeflScore_isna = data['toeflScore'].isna()
toefl_notna = data[['toefl_s', 'toefl_l', 'toefl_w', 'toefl_r']].notna().any(axis=1)
rows_to_update = toeflScore_isna & toefl_notna
data.loc[rows_to_update, 'toeflScore'] = data.loc[rows_to_update, ['toefl_s', 'toefl_l', 'toefl_w', 'toefl_r']].sum(axis=1)

data.drop(columns=['toeflEssay','toefl_s', 'toefl_l', 'toefl_w', 'toefl_r'], inplace=True)

data['ielts_score'] = data[['ielts_s', 'ielts_l', 'ielts_w', 'ielts_r']].sum(axis=1)

non_zero_ielts_scores = data.loc[data['ielts_score'] != 0, 'ielts_score']
data.loc[data['ielts_score'] != 0, 'ielts_score'] = non_zero_ielts_scores / 4
data = data.drop(["ielts_s", "ielts_l", "ielts_w", "ielts_r"], axis=1)

def ielts_to_toefl(ielts_score):
    ielts_to_toefl_table = {
        9: 118,
        8.5: 115,
        8: 110,
        7.5: 102,
        7: 94,
        6.5: 79,
        6: 60,
        5.5: 46,
        5: 35,
        4.5: 23,
        4: 12,
        3.5: 0,
        3: 0
    }
    return ielts_to_toefl_table.get(ielts_score, 0)

data['toefl_score'] = data['ielts_score'].apply(ielts_to_toefl)

data['toeflScore'] = np.where(data['toeflScore'].isnull(), data['toefl_score'], data['toeflScore'])

data = data.drop('toefl_score', axis=1)

data.loc[data['toeflScore'] > 120, 'toeflScore'] = 0

data['toeflScore'].replace(0, np.nan, inplace=True)
mean_score = data['toeflScore'].mean()
data['toeflScore'].fillna(mean_score, inplace=True)

data.head()
```

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.neighbors import NearestNeighbors
import numpy as np
from joblib import Parallel, delayed
from sklearn.model_selection import train_test_split

#Computer Science
df_filtered = scaled_df[scaled_df['Major'] == 'Computer Science']

# print(df_filtered['University'].value_counts())

def filter50plus(x):
    top_accepted_df = x.copy()
    univ_counts = top_accepted_df['University'].value_counts()
    top_univs = univ_counts[univ_counts > 100].index
    top_accepted_df = top_accepted_df[top_accepted_df['University'].isin(top_univs)]
    return top_accepted_df

df_filtered = filter50plus(df_filtered)

print((df_filtered['University'].nunique()), len(df_filtered['University']))
# print(len(df_filtered))

from sklearn.model_selection import train_test_split
X = df_filtered.drop(["Major", "University"], axis=1)
y = df_filtered["University"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)

weights = {
    "CGPA": 0.3,
    "GRE_V": 0.1,
    "GRE_Q": 0.2,
    "GRE_AWA": 0.05,
    "toeflScore": 0.15,
    "industryExp": 0.1,
    "researchExp": 0.1
}

# weights = {
#     "CGPA": 1/7,
#     "GRE_V": 1/7,
#     "GRE_Q": 1/7,
#     "GRE_AWA": 1/7,
#     "toeflScore": 1/7,
#     "industryExp": 1/7,
#     "researchExp": 1/7
# }


def run_knn(weights, data, test_data):
    data_scaled = data.copy()
    for feature in weights:
        data_scaled[feature] = data_scaled[feature] * weights[feature]

    k = 15 # number of nearest neighbors
    model = NearestNeighbors(n_neighbors=k+1)  
    model.fit(data_scaled)


    test_data_scaled = test_data.copy()
    for feature in weights:
        test_data_scaled[feature] = test_data_scaled[feature] * weights[feature]
    distances, indices = model.kneighbors(test_data_scaled)

    universities = []
    for i in range(len(test_data)):
        row_indices = indices[i]  # exclude the first neighbor (itself)
        if len(row_indices) < k:
            # Skip this test point if there are less than k neighbors
            universities.append([])
            continue
        row_distances = distances[i]  # exclude the first neighbor (itself)
        #print("distances:", row_distances)
        #print("indices:", row_indices)
        weighted_distances = 1 / (row_distances ** 2)  # apply weights to the distances
        weighted_distances[np.isinf(weighted_distances)] = np.max(weighted_distances[np.isfinite(weighted_distances)]) + 1000
        weighted_indices = [(j, weighted_distances[np.where(row_indices == j)[0][0]]) for j in row_indices]
        #print("Weighted indices:", weighted_indices,)
        university_scores = {}
        for j, w in weighted_indices:
            university = scaled_df.loc[j, "University"]
            if university in university_scores:
                university_scores[university] += w
            else:
                university_scores[university] = w
        top_universities = sorted(university_scores.items(), key=lambda x: x[1], reverse=True)[:10]
        top_universities = [u[0] for u in top_universities]
        universities.append(top_universities)
    # Return the recommended universities
    return distances, indices, universities

distances, indices, results = run_knn(weights, X_train, X_test)

def validate(y_test, results):
    if len(y_test) != len(results):
        return "Length of outputs is not equal"
    correct = 0.0
    for i, val in enumerate(y_test):
        if val in results[i]:
            correct += 1
    return correct/float(len(y_test))

validation_score = validate(y_test, results)
print("Validation Score: ", validation_score)
```
